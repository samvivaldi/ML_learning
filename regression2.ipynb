```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data/click.csv', delimiter=',', skiprows=1, dtype=np.int16) #클릭횟수별 광고료
train_x = data[:,0]  
train_y = data[:,1] 
```

```python
# 표준화
mu = train_x.mean()
sigma = train_x.std()
def standardize(x):
    return (x - mu) / sigma
```

```python
train_z = standardize(train_x)
```

```python
theta = np.random.rand(3)
```

```python
theta.shape
```

```python
def to_matrix(x):
    return np.vstack([np.ones(x.size), x, x ** 2]).T
  
```

```python
X = to_matrix(train_z)
X.shape
```

```python
#hypothesis function
def h(x):
    return np.dot(x, theta)
```

```python
# 목적함수
def E(x, y):
    return 0.5 * np.sum((y - h(x)) ** 2)
```

```python
# 학습률
ETA = 1e-3
```

```python
error = E(X, train_y)
diff = 1
count = 0 # 매개변수 갱신횟수

while diff > 1e-2:
    # 갱신 결과를 임시변수에 저장한다
    theta = theta - ETA * np.dot(h(X) - train_y, X)

    # 이전 회의 오차와의 차분을 계산한다
    current_error = E(X, train_y)
    diff = error - current_error
    error = current_error

    # 로그를 출력한다
    count += 1
    log = '{}회째: theta = {}, 차분 = {:.4f}'
    print(log.format(count, theta, diff))
```

```python
x = np.linspace(-3, 3, 100)
plt.plot(train_z, train_y, 'o')
plt.plot(x, h(to_matrix(x)))
plt.show()
```
